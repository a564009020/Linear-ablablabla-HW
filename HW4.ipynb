```python
%matplotlib inline

import numpy as np
import sympy as sp
import random as rd
import math
import matplotlib.pyplot as plt
```

```python
def func(w):
    x = sp.Symbol('x')
    equ = -3*x**5 - 10*x**4 + 2000*x**3 - 6000*x**2
    return equ.subs(x, w)

def dfunc(w):
    x = sp.Symbol('x')
    diff = sp.diff(-3*x**5 - 10*x**4 + 2000*x**3 - 6000*x**2, x)
    return diff.subs(x, w)

def GD(x_start, df, epochs, lr):
    xs = np.zeros(epochs+1)
    x = x_start
    xs[0] = x
    for i in range(epochs):
        dx = df(x)
        v = -dx*lr
        x += v
        xs[i+1] = x
    return xs
```

```python
minimum = -30
for s in range(5):
    x_start = rd.randint(-30, 30)
    epochs = 100
    lr = 0.000001
    w = GD(x_start, dfunc, epochs, lr=lr)
    if(math.isinf(w[epochs]) or math.isnan(w[epochs])):
        continue
    print(w)
    if func(w[epochs]) < func(minimum):
        minimum = w[epochs]
        
    color = 'g'
    x = np.linspace(-30, 30, 2000)
    y = -3*x**5 - 10*x**4 + 2000*x**3 - 6000*x**2
    plt.plot(x, y, 'r')
    w2 = np.zeros(epochs+1)
    for i in range(epochs+1):
        w2[i] = func(w[i])
    plt.plot(w, w2, c = color, label = 'lr = {}'.format(lr))
    plt.scatter(w, w2, c=color, )
    plt.legend()
    plt.show()

print("when x = ", minimum, "has the minimum")
```
